### Covariance matrix and orthogonal projections

The sample variance measures the spread of data around the mean in one dimension:



To measure the spread of data in more than one dimension, the notion of variance can be extended to multiple dimensions:



In the case of data with 2 input features we obtain a $2 \times 2$ matrix, called the covariance matrix, which measures the linear relationship among all dimensions of data:



Intuitively, the covariance between X and Y indicates how the values of X and Y move relative to each other. If large values of X tend to happen with large values of Y, then $(X  \mathbb{E}(X))(Y  \mathbb{E}(Y))$ is positive on average and the covariance is positive. In contrast, if X tends to be small when Y is large, then $(X  \mathbb{E}(X))(Y   \mathbb{E}(Y))$ is negative on average and the covariance is negative:
