```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import sys
import seaborn as sns
```

```python
frame = pd.read_csv("voti.dat", names = ["tutto"])
```

```python
voto = []
for string in frame["tutto"]:
    if string[-2] != '\t':
        string = string[-2]+string[-1]
    else:
        string = string[-1]
    voto.append(int(string))
    
print (voto)
```

```python
from scipy.stats import norm


help(sns.axisgrid)
sns.distplot(np.array(voto), fit=norm, kde=True, color="g", kde_kws={"shade": True})
```

```python
sns.distplot(voto, bins=20, kde=False, rug=True);
```

```python
freq = [0]*32
voto = np.array(voto)
for i in range(len(np.unique(voto))):
    freq[i] = list(voto).count(i)
        
x = []
for n in range(32):
    x.append(n)
y = freq
```

```python
plt.bar(x, y)
```
